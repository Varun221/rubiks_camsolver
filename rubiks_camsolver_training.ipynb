```
# labels to detect
labels_detect = ['cube']
f = open("data/labels.names", "w+")
for lab in labels_detect:
    f.write(f"{lab}\n")
f.close()
```

```
# this will generate data/data_train.txt
# data_train.txt contains the bounding boxes data in the following format 
# path to image xmin ymin xmax ymax class
from xml.etree import ElementTree as et
import os

parent_dir = "data"
label_dir = os.path.join(parent_dir, "labels")

f = open("data/data_train.txt", "w+")

for labfile in os.listdir(label_dir):
    if labfile[0] != '.':
        filename, _ = labfile.split('.')
        tree = et.parse(os.path.join(label_dir, labfile))
        root = tree.getroot()
        for member in root.findall('object'):
                    xmin = int(member[4][0].text)
                    ymin = int(member[4][1].text)
                    xmax = int(member[4][2].text)
                    ymax = int(member[4][3].text)
                    code = labels_detect.index(member[0].text)
                    plug_str = f"/content/{os.path.join(parent_dir, 'images', filename+'.jpg')}"
                    plug_str += f" {xmin},{ymin},{xmax},{ymax},{code}\n"
                    f.write(plug_str)

f.close()

                
    
```
